```python
import mysql.connector
import random
from faker import Faker
from datetime import datetime, timedelta
```

```python
# Initialize Faker for generating random data
fake = Faker()
```

```python
# MySQL Database Connection Details
DB_CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": "root",
    "database": "soil_management"
}
```

```python
pip install mysql-connector-python
```

```python
# Connect to MySQL
conn = mysql.connector.connect(**DB_CONFIG)
cursor = conn.cursor()
```

```python
# Assuming you already have a MySQL connection and cursor (conn, cursor)

cursor.execute("""
CREATE TABLE IF NOT EXISTS soil_health (
    id INT AUTO_INCREMENT PRIMARY KEY,
    farm_location VARCHAR(255) NOT NULL,
    test_date DATE,
    nitrogen_level FLOAT,
    phosphorus_level FLOAT,
    potassium_level FLOAT,
    pH_level FLOAT,
    moisture_content FLOAT,
    irrigation_method VARCHAR(255)
)
""")

# Commit the changes to the database
conn.commit()

print("Soil table created successfully (if not exists).")
```

```python
# Function to generate random soil data
def generate_soil_data():
    # Predefined list of irrigation methods
    irrigation_methods = ["Drip", "Sprinkler", "Flood", "Surface", "Subsurface"]
    
    farm_location = fake.city()
    test_date = fake.date_between(start_date="-2y", end_date="today")
    nitrogen = round(random.uniform(0.1, 5.0), 2)  # Nitrogen level in mg/kg
    phosphorus = round(random.uniform(0.1, 5.0), 2)  # Phosphorus level in mg/kg
    potassium = round(random.uniform(0.1, 5.0), 2)  # Potassium level in mg/kg
    pH = round(random.uniform(4.5, 8.5), 2)  # pH level
    moisture = round(random.uniform(5.0, 50.0), 2)  # Moisture content in %
    irrigation_method = random.choice(irrigation_methods)  # Randomly select irrigation method
    
    return (farm_location, test_date, nitrogen, phosphorus, potassium, pH, moisture, irrigation_method)
```

```python
# Batch insert records in chunks
batch_size = 10000  # Insert 10,000 at a time
total_records = 2000000

for i in range(0, total_records, batch_size):
    data_batch = [generate_soil_data() for _ in range(batch_size)]
    
    # Execute batch insert with irrigation_method
    cursor.executemany("""
        INSERT INTO soil_health (farm_location, test_date, nitrogen_level, phosphorus_level, potassium_level, pH_level, moisture_content, irrigation_method)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """, data_batch)
    conn.commit()
    
    print(f"{i + batch_size} records inserted...")

print("Soil health data insertion completed successfully! ✅")
```
