```python
# Calculate depreciation for each vehicle
df["Depreciation"] = df["Present_Price"] - df["Selling_Price"]

# Find the most depreciated vehicle (highest depreciation)
most_depreciated = df.loc[df["Depreciation"].idxmax()]

# Find the least depreciated vehicle (lowest depreciation)
least_depreciated = df.loc[df["Depreciation"].idxmin()]

most_depreciated[["Car_Name", "Year", "Present_Price", "Selling_Price", "Depreciation"]], \
least_depreciated[["Car_Name", "Year", "Present_Price", "Selling_Price", "Depreciation"]]
```

```python
# Extract brand names from the "Car_Name" column (assuming the first word represents the brand)
df["Brand"] = df["Car_Name"].apply(lambda x: x.split()[0])

# Calculate depreciation percentage
df["Depreciation"] = ((df["Present_Price"] - df["Selling_Price"]) / df["Present_Price"]) * 100

# Compute average depreciation per brand
brand_depreciation = df.groupby("Brand")["Depreciation"].mean().sort_values()

# Display brands with the least depreciation (top 5)
brand_depreciation.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_year = df["Selling_Price"].corr(df["Year"])
correlation_kms = df["Selling_Price"].corr(df["Kms_Driven"])

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=df["Year"], y=df["Selling_Price"], alpha=0.7)
plt.title(f"Selling Price vs. Year\nCorrelation: {correlation_year:.2f}")
plt.xlabel("Manufacturing Year")
plt.ylabel("Selling Price")
plt.subplot(1, 2, 2)
sns.scatterplot(x=df["Kms_Driven"], y=df["Selling_Price"], alpha=0.7)
plt.title(f"Selling Price vs. Kms Driven\nCorrelation: {correlation_kms:.2f}")
plt.xlabel("Kms Driven")
plt.ylabel("Selling Price")

plt.tight_layout()
plt.show()
correlation_year, correlation_kms

```
