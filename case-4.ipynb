```python
df = df.drop(df[df['Total'] == 0].index)
```

```python
#Renaming the 'Type' which has almost same meaning
df.replace('Bankruptcy or Sudden change in Economic', 
           'Bankruptcy or Sudden change in Economic Status', inplace=True)
df.replace('By Other means (please specify)', 'By Other means', inplace=True)
df.replace('Not having Children(Barrenness/Impotency',
           'Not having Children (Barrenness/Impotency)', inplace=True)
```

```python
df = df.set_index(pd.Index([i for i in range(len(df))]))
df
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
#We use rc i.e a runtime configuration (rc) containing the default styles for every plot element you create
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Q.1 Yearwise Total Number Of Deaths
#Creating a series of total suicides according to year

yearwise_total_suicide = df.groupby("Year")["Total"].sum()
print('Total number of suicides in each year: ', yearwise_total_suicide)

#plotting this in a line graph
yearwise_total_suicide.plot(kind="line", figsize = (9,5), color = 'orange')
plt.title('Yearwise Total Suicides')
plt.xlabel('Year')
plt.ylabel('Number of suicides');
plt.show()
```

#As we can see the number of suicides has been increasing from the 2005 to 2011

```python

statewise_suicides = df.groupby("State")["Total"].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
statewise_suicides.plot(kind="bar", color="red")
plt.xlabel("State")
plt.ylabel("Total Suicides (2001-2012)")
plt.title("Total Suicides in India (2001-2012) - State Wise")
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

```python
#Creating a series of total suicides on the basis of states
statewise_total_suicide = df.groupby("State")["Total"].sum()
print('Total number of suicides in each State : ', statewise_total_suicide)

#plotting the above data
filter_state = pd.DataFrame(statewise_total_suicide.reset_index()) 
sort_state = filter_state.sort_values('Total',ascending = True)
f, ax = plt.subplots(figsize=(15, 4))
#sns.barplot(y = 'Total', x = 'State',data = sort_state,edgecolor = '0.5')
sns.barplot(y = 'Total', x = 'State',data = sort_state)
plt.xticks(rotation=75);
plt.show()
```

#As we can see maharashtra has the largest number of sucide case

```python
#Filtering the data according to 'type_code','Total' and 'Year'
filter_type_code = pd.DataFrame(df.groupby(["Type_code","Year"])["Total"].sum()).reset_index()

#Plotting the above data
sns.catplot(x="Type_code", y="Total",hue="Year", kind="bar", data=filter_type_code,height=7.5, aspect=10/7.5);
plt.title('Suicide According To Causes In Each Year');
plt.show()
```

```python
# Aggregate suicide numbers by gender
genderwise_suicides = df.groupby("Gender")["Total"].sum()
plt.figure(figsize=(7, 7))
plt.pie(genderwise_suicides, labels=genderwise_suicides.index, autopct='%1.1f%%', colors=["pink", "lightblue"], startangle=140)
plt.title("Suicides in India (2001-2012) by Gender")
plt.show()
```

#we can see that suicide case for male gender is more than female

```python
# Filter data for suicides based on educational background
education_suicides = df[df["Type_code"] == "Education_Status"]
education_genderwise_suicides = education_suicides.groupby(["Type", "Gender"])["Total"].sum().unstack()
education_genderwise_suicides.plot(kind="bar", figsize=(12, 6), color=["pink","lightblue"])
plt.xlabel("Educational Background")
plt.ylabel("Total Suicides")
plt.title("Suicides in India (2001-2012) by Educational Background and Gender")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

#middle,primary and pre-university have ,ore sucide tan diploma, graduate and post graduate and above

```python
# Filter data for suicides based on educational background
education_suicides = df[df["Type_code"] == "Social_Status"]
education_genderwise_suicides = education_suicides.groupby(["Type", "Gender"])["Total"].sum().unstack()
education_genderwise_suicides.plot(kind="bar", figsize=(12, 6), color=["pink","lightblue"])
plt.xlabel("Educational Background")
plt.ylabel("Total Suicides")
plt.title("Suicides in India (2001-2012) by Educational Background and Gender")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```

#As we can see the married status is having more sucide than other also the divorcee thas the least count for sucide case

```python
#Q.sucide by age group
#Age Distributiion
statewise_Age_df = df.drop(df[df.Age_group == '0-100+'].index)
counts = statewise_Age_df['Age_group'].value_counts().sort_index()
print(counts)
counts.plot(kind='bar', title='Age Distribution',figsize=(7,5),color="indigo")
plt.xticks(rotation=0)
plt.show()
```

#15-29 has the more sucide rate than other
