```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'tips' is a DataFrame
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std, color='orange')

# Show the plot
plt.show()
```

```python
sns.countplot(x='sex',data=tips,color='skyblue')
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="day", palette='rainbow')
plt.show()
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
plt.show()
```

```python

sns.violinplot(x="day", y="total_bill", data=tips, hue="day", palette='rainbow', legend=False)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for the correlation matrix
numeric_cols = tips.select_dtypes(include=['number'])

# Create the heatmap of the correlation matrix for numeric columns
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Show the plot
plt.show()
```
